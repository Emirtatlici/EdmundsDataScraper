### We can only scrape 476 pages for each filter because of the restrictions that edmunds.com set, for that reason I am going to overcome this problem by using multiple scraper filters,These are current best filters that you can use to scrape whole database.If you inspect the structure of the website you can adjust the filters on your own. 

```python
import pandas as pd
from edmunds_scraper import EdmundsDataScraper
pd.set_option("display.max_column",None)
```

```python
#FILTER 1
scraper_1 = EdmundsDataScraper(year1=2022,year2=2024,makes=["ford","chevrolet"],page_number=366)
raw_data_v1=scraper_1.scrape_data()
```

```python
clean_data_1=scraper_1.clean_data(raw_data_v1)
```
