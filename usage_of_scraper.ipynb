```python
import pandas as pd
from edmunds_scraper import EdmundsDataScraper
pd.set_option("display.max_column",None)
```

```python
#Building the Filters.
scraper = EdmundsDataScraper(year1=2020,year2=2024,makes=["bmw","mercedes-benz","audi","honda"],page_number=6)
```

```python
#Scraping the Data
raw_data=scraper.scrape_data()
clean_data=scraper.clean_data(raw_data)
```

```python
clean_data.drop(columns=["vid","vin"],inplace=True)
```
